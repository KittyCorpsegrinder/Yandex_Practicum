## Предобработка

<div style="border:solid green 2px; padding: 20px">
Для начала, приведем названия столбцов к нижнему регистру

```python
df.columns = map(str.lower, df.columns)
df.head()
```

<div style="border:solid green 2px; padding: 20px">
Теперь преобразуем данные в нужные типы.
    В столбце years_of_release необходимо изменить данные из дробных в целые, т.к. это год, и логично его выводить целым значением. na, eu, jp_sales есть смысл оставить значениями с плавающей точкой, ведь мы знаем что это продажи в миллионах и так удобнее их воспринимать. Оценки критиков и пользователей стоит оставить такими же значениями, потому что это среднее или медианное значение взятое с других ресурсов где пользователи ставили оценку, а потому числа целыми быть не могут (мое предположение).

```python
# Заполняем пропуски в столбце года релиза маркером 2020
df['year_of_release'] = df['year_of_release'].fillna(2020)
df['year_of_release'].isna().sum()

# Переводим значения в целые числа
df['year_of_release'] = df['year_of_release'].astype('int')
#df.head()
```

<div style="border:solid green 2px; padding: 20px">
Я удалил пропуски в столбце даты релиза потому что их невозможно заменить на среднее или медианное значение, в силу того что это исказит значения при выводе графика продаж по временам года, например. Считаю правильным удалить эти значения, не искажая данные.  
    

    
    Пропуски могли появиться при повреждении датафрейма, так же этому может способствовать человеческий фактор.

```python
# В столбцах genre и name есть по два пропуска. Считаю их следует удалить
df.dropna(subset=['genre', 'name'], inplace=True)
df[['genre', 'name']].isna().sum()

# Проверим уникальные значения столбцов platform, genre
df['genre'].unique() # Никаких странных названий, значений, все хорошо
df['platform'].unique() # Какие-то странные названия некоторых платформ (ну скорее всего это я не шарю в разновидностях консолей)

# Дубликаты в столбцах name
df.drop_duplicates(subset=['name']).head()
```

<div style="border:solid green 2px; padding: 20px">
Теперь возьмемся за столбцы critic_score и user_score. Как известно, это оценка критиков, и отдельно оценка пользователей, сначала выведем процентное соотношение пропусков относительно заполненных данных, затем уникальные значения чтобы понимать что содержится в столбцах, а после поглядим что можно с этим сделать.

```python
# Выводим процентное соотношение пропущенных значений
for col in df.columns:
    print(f'{col}:{df[col].isna().sum() / df.shape[0]*100:.2f}%')
```

<div style="border:solid green 2px; padding: 20px">
Ну... 51% и 40% слишком большой объем пропусков для варварского удаления из датафрейма, пока оставлю все так как оно есть

```python
# Уникальные значения 'critic_score'
df['critic_score'].unique() # Ничего странного
df['critic_score'].describe()

# Теперь 'user_score'
df['user_score'].unique() # Видим странное значение 'tbd'
df['user_score'].describe() # 'tbd' и столбец в object. Требуется исправить

# Пока не буду трогать пропуски, т.к. не совсем понимаю как они повлияют в дальнейшем на работу
```

<div style="border:solid green 2px; padding: 20px">
TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. Акроним служит для отметки неясностей или пропусков, которые надо заполнить, в информации требований.
    
    Следовательно, стоит заменить tbd на пропуски, а затем привести столбец к float

```python
df.user_score.replace('tbd', np.nan, inplace=True)
df['user_score'] = df['user_score'].astype('float')
df['user_score'].describe()
```

<div style="border:solid green 2px; padding: 20px">
Остался столбец с рейтингом, который следует изучить.

```python
# Откроем уникальыне значения рейтинга
df['rating'].unique()
```

<div style="border:solid green 2px; padding: 20px">
Расшифровка рейтингов.   
    
-«EC» («Early childhood») — «Для детей младшего возраста»  
-«E» («Everyone») — «Для всех»  
-«E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»  
-«T» («Teen») — «Подросткам»  
-«M» («Mature») — «Для взрослых»  
-«AO» («Adults Only 18+») — «Только для взрослых»      
-«RP» («Rating Pending») — «Рейтинг ожидается»  
-«RP 17+» («Rating Pending») — «Рейтинг ожидается - примерно 17+»  
    
    Важный момент, что ESRB была основана в 1994 году.

```python
# Исходя из расшифровки, следует что К-А ошибочное значение, глянем какую часть данных оно занимает
df['rating'].hist(bins=10);
```

<div style="border:solid green 2px; padding: 20px">
Скорее всего, обозначение К-А появилось из-за человеческого фактора, и я просто заменю его на E (для всех).  
Пропуски в столбце рейтинга можно допустить за норму, т.к. я указывал выше, что важный момент дата основания организации ESRB. До 1994 года разделение игр на рейтинги не существовало.

```python
# Заменяем К-А
df.loc[df['rating'] == 'K-A', 'rating'] = 'E'

# Посмотрим на график рейтинга выстроенного по шкале времени
rating = df.groupby('year_of_release')['rating'].count().reset_index().plot(x='year_of_release', grid=True)
plt.title('Рейтинг игр по ESRB')
plt.xlabel('Год')
plt.ylabel('Игры')
plt.show()

# Пропуски не удаляем, заменю маркером 'unknown'
df['rating'] = df['rating'].fillna('unknown')
df['rating'].unique()
```

<div style="border:solid green 2px; padding: 20px">
На графике видно что с 1980 до 1995 год линия имеет нулевое значение. А значит, игры не подвергались сортировке по рейтингу, и мое предположение оказалось верным.

<div style="border:solid green 2px; padding: 20px">
Создадим новый столбец sum_sales, в котором будет сумма продаж по трем регионам вместе.

```python
df['sum_sales'] = df.apply(lambda row: row['na_sales'] + row['eu_sales'] + row['jp_sales'] + row['other_sales'], axis=1)
df.head()
```

<div style="border:solid green 2px; padding: 20px">
Можно сделать небольшой вывод о проделанном. Данные были приведены к соответствующим типам, названия столбцов были приведены к нижнему регистру, создан отдельный столбец с общей суммой продаж в разных регионах, удалены дубликаты и пропуски там где это оказалось необходимым. ## Проверка гипотез

```python
actual_years['platform'].isna().sum()
```

<div style="border:solid green 2px; padding: 20px">
    
    
- Нулевая: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.      
- Первая: Средние пользовательские рейтинги жанров Action и Sports разные.

```python
# Удаляем пропуски, получается...
actual_years.dropna(axis=0, inplace=True)
actual_years.isna().sum()
```

```python
#выберем пользовательские рейтинги платформ Xbox One 
user_score_xone= actual_years.query('platform == "XOne" and user_score!="nan" and year_of_release >= 2012') 
#выберем пользовательские рейтинги платформ Xbox One и PC
user_score_pc = actual_years.query('platform == "PC" and user_score!="nan" and year_of_release >= 2012')  

var_xone = np.var(user_score_xone['user_score'])
var_pc= np.var(user_score_pc['user_score'])
print('{:} {: >5.2f} '.format('Дисперсия платформ Xbox One равна:', var_xone))
print('{:} {: >11.2f} '.format('Дисперсия платформ PC равна:', var_pc ))

alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(
    user_score_xone['user_score'],
    user_score_pc['user_score'],
    equal_var= False  # так как дисперсии генеральных совокупностей, из которых взяты выборки, не равны
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, т.е. средние пользовательские рейтинги платформ Xbox One и PC различаются!")
else:
    print("Не получилось отвергнуть нулевую гипотезу, средние пользовательские рейтинги платформ Xbox One и PC одинаковы!")
```

<div style="border:solid green 2px; padding: 20px">
Таким образом, видим что нулевая гипотеза о равности средних пользовательских рейтингах платформ XOne и PC не отвергается. p-значение в 0.18 не является достаточным основанием для того чтобы отвергнуть гипотезу.

```python
#выберем пользовательские рейтинги жанров Action и Sports
user_score_action= actual_years.query('genre == "Action" and user_score!="nan" and year_of_release >= 2012') 
#выберем пользовательские рейтинги платформ Xbox One и PC
user_score_sports = actual_years.query('genre == "Sports" and user_score!="nan" and year_of_release >= 2012')  

var_action = np.var(user_score_action['user_score'])
var_sports= np.var(user_score_sports['user_score'])
print('{:} {: >5.2f} '.format('Дисперсия жанра Action равна:', var_action))
print('{:} {: >5.2f} '.format('Дисперсия жанра Sports равна:', var_sports))

alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(
    user_score_action['user_score'],
    user_score_sports['user_score'],
    equal_var= False  # так как дисперсии генеральных совокупностей, из которых взяты выборки, не равны
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, т.е. cредние пользовательские рейтинги жанров Action и Sport разные!")
else:
    print("Не получилось отвергнуть нулевую гипотезу, т.е cредние пользовательские рейтинги жанров Action и Sports одинаковые!")
```

<div style="border:solid green 2px; padding: 20px">
р-значение больше единицы, а значит нулевую гипотезу можно отвергнуть. Средние пользовательские рейтинки жанров экшн  и спорт разные.
